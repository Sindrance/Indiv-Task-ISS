```python
import requests
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import time
%matplotlib inline

from mpl_toolkits import mplot3d
```

```python
    r = requests.get('http://api.open-notify.org/iss-now.json')
    space_station_location = (r.json())
    
    lat_list=[]
    long_list=[]
    for i in range(101):
        r = requests.get(url='http://api.open-notify.org/iss-now.json')
        space_station_location = (r.json())
        
        space_station_latitude = float(space_station_location ['iss_position']['latitude'])
        space_station_longitude = float(space_station_location['iss_position']['longitude'])
        
        lat_list.append(space_station_latitude)
        long_list.append(space_station_longitude)
        time.sleep(10)
        
    print(lat_list)
        


```

```python
fig = plt.figure()
fig.set_size_inches(15, 15)

plt.title("Position of the Internation Space Station")
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.grid()

plt.scatter(lat_list,long_list, c="blue")
plt.show()
```
